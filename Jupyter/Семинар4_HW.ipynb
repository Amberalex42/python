## Задача 3.

Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
plt.figure(figsize=(10, 6))
corr_matrix = df.corr()
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm');
```

Исходя из тепловой карты, можно сделать вывод, что цена дома имеет тесную прямую связь с площадью жилой части (коэффициент корреляции = 0.7), и с рейтингом (коэффициент корреляции = 0.7). Также сильная связь есть с площадью надземной части дома и количеством ванных комнат.

```python
plt.figure(figsize=(8, 6))
sns.boxplot(y=df['price'], x=df['floors'], whis=1.5)
plt.title('Зависимость цены от этажности')
plt.xlabel('Количество этажей')
plt.ylabel('Цена');
```

Средняя цена выше всего у домов, имеющих 2.5 этажа, также у них довольно низкая степень выбросов.

```python
plt.figure(figsize=(8, 6))
sns.boxplot(y=df['price'], x=df['bathrooms'], whis=1.5)
plt.title('Зависимость цены от количества ванных комнат')
plt.xlabel('Количество ванных комнат')
plt.ylabel('Цена')
plt.xticks(rotation=60);
```

Наблюдается прямая закономерность: чем больше число ванных комнат, тем выше стоимость дома.

```python
plt.figure(figsize=(8, 6))
sns.boxplot(y=df['price'], x=df['grade'], whis=1.5)
plt.title('Зависимость цены от оценки')
plt.xlabel('Оценка')
plt.ylabel('Цена')
plt.xticks(rotation=60);
```

Также наблюдается прямая закономерность: чем выше оценка, тем выше цена квартиры.

```python
plt.figure(figsize=(8, 6))
sns.boxplot(y=df['price'], x=df['condition'], whis=1.5)
plt.title('Зависимость цены от состояния')
plt.xlabel('Состояние дома')
plt.ylabel('Цена')
plt.xticks(rotation=60);
```
